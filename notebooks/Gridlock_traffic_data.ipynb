```python
df['cargo_material'] = df['cargo_material'].fillna('unknown').astype(str).str.lower().str.strip()

# Step 2: Define a robust keyword-matching function
def categorize_cargo(text):
    # Check for empty or non-applicable trucks
    if any(word in text for word in ['empty', 'emty', 'nil', 'no', 'unknown']):
        return 'empty_or_none'
    # Check for heavy construction materials
    elif any(word in text for word in ['brick', 'mixer', 'kambi', 'mud', 'electric']):
        return 'construction_heavy'
    # Check for garbage/municipal vehicles
    elif any(word in text for word in ['garbage', 'bbmp']):
        return 'municipal_waste'
    # Check for liquids or hazardous gas
    elif any(word in text for word in ['milk', 'gas']):
        return 'liquid_gas'
    # Check for food/perishables
    elif any(word in text for word in ['food', 'grocer', 'poultry', 'cattle']):
        return 'perishables'
    # Catch-all for general goods, Amazon items, and Kannada entries
    else:
        return 'general_goods'

# Apply the function to create our clean macro column
df['cargo_macro'] = df['cargo_material'].apply(categorize_cargo)

# Step 3: Safely drop the original chaotic column
df = df.drop(columns=['cargo_material'])

# Step 4: Apply One-Hot Encoding to the clean macro column
# drop_first=True prevents the dummy variable trap
df = pd.get_dummies(df, columns=['cargo_macro'], drop_first=True, dtype=int)

# Verify the result by looking at the new encoded columns
print([col for col in df.columns if 'cargo_macro' in col])
```

```python
df['reason_breakdown'] = df['reason_breakdown'].fillna('not_applicable').astype(str).str.lower().str.strip()

def categorize_breakdown(text):
    # Base case: if it wasn't a breakdown
    if text == 'not_applicable':
        return 'not_applicable'

    # 1. Minor Issues: Fuel, Battery, Tires (Includes Kannada keywords for puncture/tyre)
    if any(word in text for word in ['diesel', 'fuel', 'empty', 'emty', 'battery', 'battiry', 'puncture', 'puncher', 'tyre', 'ಟೈರ್', 'ಪಂಚರ್']):
        return 'minor_issue'

    # 2. Major Failures: Axles, Gears, Air Locks, Radiator, Clutch
    elif any(word in text for word in ['gear', 'axil', 'accel', 'clutch', 'belt', 'radiator', 'air', 'aer', 'wire', 'boat', 'jain']):
        return 'major_failure'

    # 3. General/Vague Breakdowns (Catch-all for 'starting problem', 'breakdown', etc.)
    else:
        return 'general_breakdown'

df['breakdown_severity'] = df['reason_breakdown'].apply(categorize_breakdown)

df = df.drop(columns=['reason_breakdown'])


df = pd.get_dummies(df, columns=['breakdown_severity'], drop_first=True, dtype=int)
```

```python
from sklearn.model_selection import train_test_split


X = df.drop(columns=['congestion_time_minutes', 'requires_road_closure'])

y_duration = df['congestion_time_minutes']
y_closure = df['requires_road_closure']



X_train_dur, X_test_dur, y_train_dur, y_test_dur = train_test_split(
    X, y_duration, test_size=0.2, random_state=42
)



X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X, y_closure, test_size=0.2, random_state=42, stratify=y_closure
)

print("X Training Shape:", X_train_dur.shape)
print("X Testing Shape:", X_test_dur.shape)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from xgboost import XGBRegressor, XGBClassifier
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, f1_score, roc_auc_score

# ==========================================
# 1. THE DURATION MODEL (REGRESSION)
# ==========================================
print("--- CONGESTION DURATION MODELS (REGRESSION) ---")

regression_models = {
    "Linear Regression": LinearRegression(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "XGBoost": XGBRegressor(n_estimators=100, random_state=42, objective='reg:squarederror')
}

for name, model in regression_models.items():
    # Train the model
    model.fit(X_train_dur, y_train_dur)

    # Make predictions on the unseen Test set
    y_pred = model.predict(X_test_dur)

    # Evaluate
    mae = mean_absolute_error(y_test_dur, y_pred)
    r2 = r2_score(y_test_dur, y_pred)

    print(f"\\n{name}:")
    print(f"  MAE (Mean Absolute Error): {mae:.2f} minutes")
    print(f"  R-Squared: {r2:.4f}")

print("\\n" + "="*50 + "\\n")

# ==========================================
# 2. THE LOGISTICS MODEL (CLASSIFICATION)
# ==========================================
print("--- ROAD CLOSURE MODELS (CLASSIFICATION) ---")

classification_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(n_estimators=100, random_state=42, eval_metric='logloss')
}

for name, model in classification_models.items():
    # Train the model
    model.fit(X_train_cls, y_train_cls)

    # Make predictions
    y_pred = model.predict(X_test_cls)
    y_prob = model.predict_proba(X_test_cls)[:, 1] # Get probabilities for ROC-AUC

    # Evaluate
    acc = accuracy_score(y_test_cls, y_pred)
    f1 = f1_score(y_test_cls, y_pred)
    roc = roc_auc_score(y_test_cls, y_prob)

    print(f"\\n{name}:")
    print(f"  Accuracy: {acc:.4f}")
    print(f"  F1-Score: {f1:.4f}")
    print(f"  ROC-AUC:  {roc:.4f}")
```
